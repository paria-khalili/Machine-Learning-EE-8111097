# Evaluation

Aggregate classification and clustering metrics for comparison and reporting.

```python
from pathlib import Path
import json

import pandas as pd

from src.config import PROJECT_ROOT
```

## Classification summary

```python
summary_path = PROJECT_ROOT / "data" / "processed" / "classification_summary.csv"
metrics_path = PROJECT_ROOT / "data" / "processed" / "classification_metrics.json"

metrics = {}
if metrics_path.exists():
    with metrics_path.open("r", encoding="utf-8") as handle:
        metrics = json.load(handle)

if summary_path.exists():
    summary_df = pd.read_csv(summary_path, index_col=0)
elif metrics:
    rows = {}
    for name, report in metrics.items():
        rows[name] = {
            "accuracy": report.get("accuracy"),
            "precision_weighted": report["weighted avg"]["precision"],
            "recall_weighted": report["weighted avg"]["recall"],
            "f1_weighted": report["weighted avg"]["f1-score"],
            "precision_macro": report["macro avg"]["precision"],
            "recall_macro": report["macro avg"]["recall"],
            "f1_macro": report["macro avg"]["f1-score"],
        }
    summary_df = pd.DataFrame(rows).T
else:
    summary_df = pd.DataFrame()

if not summary_df.empty:
    summary_df = summary_df.sort_values("f1_weighted", ascending=False)

summary_df
```

```python
best_model = summary_df.index[0] if not summary_df.empty else None
best_model
```

```python
if best_model and best_model in metrics:
    per_class_df = pd.DataFrame(metrics[best_model]).T
else:
    per_class_df = pd.DataFrame()

per_class_df
```
